```python
# Stacked Bar Chart for Sales Representatives by Sales Channel
sales_channel_pivot = sales_channel_summary.pivot(index='SalesRep ID', columns='Sales Channel', values='Frequency').fillna(0)

# Plotting the stacked bar chart
sales_channel_pivot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title('Sales Frequency by Sales Representative and Sales Channel')
plt.xlabel('Sales Representative ID')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Sales Channel')
plt.tight_layout()
plt.show()
```

```python
# Calculate frequency of purchases by Product Category
product_category_freq = df['Product Category'].value_counts()

# Calculate percentage of total purchases by Product Category
product_category_percentage = (product_category_freq / product_category_freq.sum()) * 100

# Combine into a DataFrame
product_category_summary = pd.DataFrame({
    'Frequency': product_category_freq,
    'Percentage (%)': product_category_percentage
}).reset_index()

product_category_summary.columns = ['Product Category', 'Frequency', 'Percentage (%)']

# Sort the product categories by frequency and select the top 5
top_5_categories = product_category_summary.nlargest(5, 'Frequency')

# Calculate "Other" category (sum of all remaining categories)
other_frequency = product_category_summary['Frequency'][5:].sum()
other_percentage = product_category_summary['Percentage (%)'][5:].sum()

# Append "Other" to the top 5 categories
# Use concat instead of append and create a new DataFrame for the 'Other' row
top_5_categories = pd.concat([top_5_categories, pd.DataFrame({'Product Category': ['Other'], 'Frequency': [other_frequency], 'Percentage (%)': [other_percentage]})], ignore_index=True)

# Pie Chart for Top 5 Product Categories
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(top_5_categories['Percentage (%)'], labels=top_5_categories['Product Category'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Top 5 Product Categories by Purchase Percentage')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## SECTION 12. PROFIT MARGIN ANALYSIS

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # For color palette

# Load the Excel file
file_path = '/content/DATA for sales cleaned.xlsx'  # Update this with your actual file path
sales_data = pd.read_excel(file_path, sheet_name='Sales')

# Calculate Profit Margin for each product
sales_data['Profit Margin'] = (sales_data['Sales Amount'] - sales_data['Cost Of Goods Sold']) / sales_data['Sales Amount'] * 100

# Extract relevant columns for analysis
profit_margin_analysis = sales_data[['Product Name', 'Cost Of Goods Sold', 'Sales Amount', 'Profit Margin']]
print(profit_margin_analysis)

# Drop duplicates to get unique products and sort by Profit Margin
profit_margin_analysis_unique = profit_margin_analysis.drop_duplicates(subset=['Product Name']).sort_values('Profit Margin', ascending=False)

# Select top 10 products by profit margin
top_10_profit_margin = profit_margin_analysis_unique.head(10)

# Bar chart: Compare top 10 profit margins across products
plt.figure(figsize=(10, 6))

# Generate a color palette with 10 different colors
colors = sns.color_palette("husl", 10)

plt.bar(top_10_profit_margin['Product Name'], top_10_profit_margin['Profit Margin'], color=colors)
plt.title('Top 10 Products by Profit Margins')
plt.xlabel('Product Name')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Scatter plot: Show relationship between profit margins and feedback frequency
# Assuming 'Feedback Frequency' is available in the dataset
feedback_data = sales_data.groupby('Product Name').size().reset_index(name='Feedback Frequency')

# Merge feedback data with profit margin analysis
merged_data = pd.merge(profit_margin_analysis_unique, feedback_data, on='Product Name')

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Profit Margin'], merged_data['Feedback Frequency'], color='green', marker='o')
plt.title('Profit Margin vs Feedback Frequency')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Feedback Frequency')
plt.tight_layout()
plt.show()
```
