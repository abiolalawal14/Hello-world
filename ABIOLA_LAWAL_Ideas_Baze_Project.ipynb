```python
# First, import the pandas library
import pandas as pd

# Define the file path and sheet name
file_path = '/content/DATA for sales cleaned.xlsx'
sheet_name = 'Sales'

# Read the specific sheet from the Excel file
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Check if the data is loaded correctly
print(df.head())
```

```python
# Count the frequency of each feedback category
feedback_freq = df['FEEDBACK ANALYSIS'].value_counts()

# Compute the percentage for each category
feedback_percent = (feedback_freq / feedback_freq.sum()) * 100

# Combine frequency and percentage into a single DataFrame for better display
feedback_summary = pd.DataFrame({
    'Frequency': feedback_freq,
    'Percentage (%)': feedback_percent
})

# Display the result
feedback_summary
```

```python
plt.figure(figsize=(8,8))
feedback_summary['Percentage (%)'].plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'coral', 'gold', 'purple', 'pink'])
plt.title('Feedback Analysis Percentage Distribution')
plt.ylabel('')  # Hide y-label for pie chart
plt.show()
```

```python
# Group by Region and calculate the total and average sales amount for each region
region_sales_summary = df.groupby('Region')['Sales Amount'].agg(
    Total_Amount='sum',  # Total sales amount
    Average_Amount='mean'  # Average sales amount
).reset_index()

# Calculate the percentage of total sales for each region
total_sales = region_sales_summary['Total_Amount'].sum()
region_sales_summary['Percentage of Total Sales (%)'] = (region_sales_summary['Total_Amount'] / total_sales) * 100

# Display the summary
region_sales_summary
```

```python
plt.figure(figsize=(10,6))
region_sales_summary.set_index('Region')['Total_Amount'].plot(kind='bar', color='cornflowerblue')
plt.title('Total Sales Amount by Region')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# Load the Excel file with the correct path and sheet name
file_path = '/content/DATA for sales cleaned.xlsx'
df = pd.read_excel(file_path, sheet_name='Sales')

# Check if the data is loaded correctly
print(df.head())
```

```python
# Print the column names to verify if 'SalesRep ID' exists
print(df.columns)
```

```python
# Calculate frequency of sales by each Sales Rep
sales_rep_freq = df['SalesRep ID'].value_counts()

# Calculate percentage of each Sales Rep's sales
sales_rep_percentage = (sales_rep_freq / sales_rep_freq.sum()) * 100

# Combine into a DataFrame
sales_rep_summary = pd.DataFrame({
    'Frequency': sales_rep_freq,
    'Percentage (%)': sales_rep_percentage
}).reset_index()

sales_rep_summary.columns = ['Sales Rep ID', 'Frequency', 'Percentage (%)']

# Display the summary
print(sales_rep_summary)
```

```python
# Calculate frequency by Sales Rep and Sales Channel
sales_channel_summary = df.groupby(['SalesRep ID', 'Sales Channel']).size().reset_index(name='Frequency')

# Calculate percentage for each Sales Rep and Sales Channel
total_sales = sales_channel_summary['Frequency'].sum()
sales_channel_summary['Percentage (%)'] = (sales_channel_summary['Frequency'] / total_sales) * 100
```

```python
import matplotlib.pyplot as plt

# Bar Chart for Sales Representatives
plt.figure(figsize=(12, 6))
plt.bar(sales_rep_summary['Sales Rep ID'], sales_rep_summary['Frequency'], color=['green', 'blue', 'red'])
plt.title('Sales Frequency by Sales Representative')
plt.xlabel('Sales Representative ID')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Stacked Bar Chart for Sales Representatives by Sales Channel
sales_channel_pivot = sales_channel_summary.pivot(index='SalesRep ID', columns='Sales Channel', values='Frequency').fillna(0)

# Plotting the stacked bar chart
sales_channel_pivot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title('Sales Frequency by Sales Representative and Sales Channel')
plt.xlabel('Sales Representative ID')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Sales Channel')
plt.tight_layout()
plt.show()
```

Insights
Identifying Top Performers:

Sales representatives with the highest frequencies are those associated with more purchases. This indicates strong sales performance, potentially due to effective sales techniques, strong relationships with customers, or superior product knowledge.
For example, if Rep_01 shows a significantly higher frequency than others, they may be excelling in their sales efforts or have a larger territory.
Sales Channel Effectiveness:

By examining the stacked bar chart, we can identify which sales channels contribute most to each representative’s success. For instance, if Rep_03 predominantly makes sales through the Wholesale channel, this may highlight a specialization that can be leveraged.
Any representatives with unusually low frequencies, despite being part of a popular sales channel, may warrant further investigation.
Strange Occurrences:

If a representative has a high percentage of sales but a very low frequency, this could indicate they are closing larger deals but not many of them. Alternatively, a rep with a high frequency but low value sales might suggest issues in converting leads to high-value sales.
Look for representatives whose performance deviates significantly from the average. For example, if Rep_05 has very few sales but a high percentage, it may indicate irregularities that need to be addressed.
Recommendations
Performance Reviews: Conduct performance reviews for top and bottom sales representatives to identify best practices and areas needing improvement.
Training and Development: Provide additional training or resources for underperforming sales representatives, focusing on strategies that have proven successful for their peers.
Sales Channel Optimization: Evaluate the effectiveness of each sales channel per representative to strategize better allocation of resources and training.
Targeted Incentives: Consider creating incentive programs based on feedback and performance data to motivate sales representatives and enhance productivity.

```python
# Calculate frequency of purchases by Product Category
product_category_freq = df['Product Category'].value_counts()

# Calculate percentage of total purchases by Product Category
product_category_percentage = (product_category_freq / product_category_freq.sum()) * 100

# Combine into a DataFrame
product_category_summary = pd.DataFrame({
    'Frequency': product_category_freq,
    'Percentage (%)': product_category_percentage
}).reset_index()

product_category_summary.columns = ['Product Category', 'Frequency', 'Percentage (%)']

# Sort the product categories by frequency and select the top 5
top_5_categories = product_category_summary.nlargest(5, 'Frequency')

# Calculate "Other" category (sum of all remaining categories)
other_frequency = product_category_summary['Frequency'][5:].sum()
other_percentage = product_category_summary['Percentage (%)'][5:].sum()

# Append "Other" to the top 5 categories
# Use concat instead of append and create a new DataFrame for the 'Other' row
top_5_categories = pd.concat([top_5_categories, pd.DataFrame({'Product Category': ['Other'], 'Frequency': [other_frequency], 'Percentage (%)': [other_percentage]})], ignore_index=True)

# Pie Chart for Top 5 Product Categories
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(top_5_categories['Percentage (%)'], labels=top_5_categories['Product Category'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Top 5 Product Categories by Purchase Percentage')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
